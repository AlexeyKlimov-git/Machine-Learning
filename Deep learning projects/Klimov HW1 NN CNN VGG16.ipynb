```python
def balanced_dataset(df):
    df_balanced = pd.DataFrame()
    #df = pd.DataFrame()
    
    for cat in df['cell_type_idx'].unique():
        temp = resample(df[df['cell_type_idx'] == cat], 
                        replace=True,     # sample with replacement
                        n_samples=7000,   # to match majority class !!! –≤–æ—Ç —Ç—É—Ç –¥–æ–ø–æ–ª–Ω—è–µ–º –æ—Å—Ç–∞–ª—å–Ω—ã–µ –∫–ª–∞—Å—Å—ã –¥–æ –æ–¥–∏–Ω–∞–∫–æ–≤–æ–≥–æ –∫–æ–ª-–≤–∞ —ç–∫–∑–µ–º–ø–ª—è—Ä–æ–≤ 
                        random_state=123) # reproducible results

        # Combine majority class with upsampled minority class
        df_balanced = pd.concat([df_balanced, temp])
 
    df_balanced['cell_type'].value_counts()

    return df_balanced
```

```python
def load_img_data(size, df, balanced=False):
    """
        ..
        first we should normalize the image from 0-255 to 0-1
    """
    
    img_h, img_w = size, size
    imgs = []
    
    if balanced:
        df = balanced_dataset(df)
    
    image_paths = list(df['path'])

    for i in tqdm(range(len(image_paths))):
        img = cv2.imread(image_paths[i])
        img = cv2.resize(img, (img_h, img_w))
        img = img.astype(np.float32) / 255.
        #img = np.asarray(Image.open(image_paths[i]).resize((size,size)))
        imgs.append(img)

    imgs = np.stack(imgs, axis=0)
    print(imgs.shape)

    #imgs = imgs.astype(np.float32) / 255.
    
    return imgs, df['cell_type_idx'].values
```

```python
import numpy as np
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
%matplotlib inline

# –º–æ–¥—É–ª–∏ –±–∏–±–ª–∏–æ—Ç–µ–∫–∏ PyTorch
import torch
from torchvision import datasets, transforms
# –º–µ—Ç—Ä–∏–∫–∞ –∫–∞—á–µ—Å—Ç–≤–∞ 
from sklearn.metrics import accuracy_score
```

```python
# –Ω–∞–º –Ω–µ –Ω—É–∂–µ–Ω –ø–æ—Å–ª–µ–¥–Ω–∏–π —Å–ª–æ–π, –æ–±—Ä–µ–∑–∞–µ–º –µ–≥–æ
vgg16.classifier = nn.Sequential(*list(vgg16.classifier.children()))[:-1]
```

```python
class New_VGG16(nn.Module):
    def __init__(self):
        super().__init__()
        self.vgg16 = vgg16 # –≤—Å—è –æ–≥—Ä–æ–º–Ω–∞—è –Ω–µ–π—Ä–æ—Å–µ—Ç—å –≤ –æ–¥–Ω—É —Å—Ç—Ä–æ—á–∫—É
        for param in self.vgg16.features.parameters(): # –æ—Ç–∫–ª—é—á–∞–µ–º –¥–ª—è –Ω–µ–µ –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–æ–≤
            param.requires_grad = False
        self.fc = nn.Linear(4096, 2) # –¥–æ–±–∞–≤–ª—è–µ–º –Ω–æ–≤—ã–π —Å–ª–æ–π
    
    def forward(self, x):
        # forward pass —Å–µ—Ç–∏
        # —É–º–Ω–æ–∂–µ–Ω–∏–µ –Ω–∞ –º–∞—Ç—Ä–∏—Ü—É –≤–µ—Å–æ–≤ 1 —Å–ª–æ—è –∏ –ø—Ä–∏–º–µ–Ω–µ–Ω–∏–µ —Ñ—É–Ω–∫—Ü–∏–∏ –∞–∫—Ç–∏–≤–∞—Ü–∏–∏
        x = self.vgg16(x)
        x = self.fc(x)
        return x
```

```python
def train(net, n_epoch=3):
    # –≤—ã–±–∏—Ä–∞–µ–º —Ñ—É–Ω–∫—Ü–∏—é –ø–æ—Ç–µ—Ä—å
    loss_fn = torch.nn.CrossEntropyLoss()

    # –≤—ã–±–∏—Ä–∞–µ–º –∞–ª–≥–æ—Ä–∏—Ç–º –æ–ø—Ç–∏–º–∏–∑–∞—Ü–∏–∏ –∏ learning_rate
    learning_rate = 1e-3
    optimizer = torch.optim.Adam(net.parameters(), lr=learning_rate)

    # acc –ø–æ test
    best_accuracy = 0
    # –æ–±—É—á–∞–µ–º —Å–µ—Ç—å 2 —ç–ø–æ—Ö–∏
    for epoch in tqdm(range(n_epoch)):

        running_loss = 0.0
        train_dataiter = iter(train_loader)
        for i, batch in enumerate(tqdm(train_dataiter)):
            # —Ç–∞–∫ –ø–æ–ª—É—á–∞–µ–º —Ç–µ–∫—É—â–∏–π –±–∞—Ç—á
            X_batch, y_batch = batch
            # –ø–µ—Ä–µ–Ω–æ—Å–∏–º –µ–≥–æ –Ω–∞ –≤–∏–¥–µ–æ–ø–∞–º—è—Ç—å
            # –µ—Å–ª–∏ —Ç–æ—á–Ω–æ —É–≤–µ—Ä–µ–Ω—ã, —á—Ç–æ —ç—Ç–æ –≥–ø—É - –º–æ–∂–Ω–æ –Ω–∞–ø–∏—Å–∞—Ç—å .cuda()
            X_batch = X_batch.to(device)
            y_batch = y_batch.to(device)
            # –æ–±–Ω—É–ª—è–µ–º –≤–µ—Å–∞
            optimizer.zero_grad()

            # forward pass (–ø–æ–ª—É—á–µ–Ω–∏–µ –æ—Ç–≤–µ—Ç–æ–≤ –Ω–∞ –±–∞—Ç—á –∫–∞—Ä—Ç–∏–Ω–æ–∫)
            y_pred = net(X_batch)
            # –≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –ª–æ—Å—Å–∞ –æ—Ç –≤—ã–¥–∞–Ω–Ω—ã—Ö —Å–µ—Ç—å—é –æ—Ç–≤–µ—Ç–æ–≤ –∏ –ø—Ä–∞–≤–∏–ª—å–Ω—ã—Ö –æ—Ç–≤–µ—Ç–æ–≤ –Ω–∞ –±–∞—Ç—á
            loss = loss_fn(y_pred, y_batch)
            # bsckpropagation (–≤—ã—á–∏—Å–ª–µ–Ω–∏–µ –≥—Ä–∞–¥–∏–µ–Ω—Ç–æ–≤)
            loss.backward()
            # –æ–±–Ω–æ–≤–ª–µ–Ω–∏–µ –≤–µ—Å–æ–≤ —Å–µ—Ç–∏
            optimizer.step()

            # –ø—Ä–æ–≤–µ—Ä—è–µ–º –∫–∞—á–µ—Å—Ç–≤–æ –∫–∞–∂–¥—ã–µ 1500 –±–∞—Ç—á–µ–π (~–ø–æ–ª —ç–ø–æ—Ö–∏)
            if i % 1500 ==0:
                # –º–µ–Ω–µ–¥–∂–µ—Ä —É–ø–∞—Ä–≤–ª–µ–Ω–∏—è –∫–æ–Ω—Ç–µ–∫—Å—Ç–æ–º —Ç–æ—Ä—á–∞ —É–∫–∞–∑—ã–≤–∞–µ—Ç –Ω–∞ —Ç–æ, —á—Ç–æ–±—ã –Ω–µ –æ–±–Ω–æ–≤–ª—è—Ç—å –ø–∞—Ä–∞–º–µ—Ç—Ä—ã
                with torch.no_grad():
                    accuracy = []
                    for batch in test_loader:
                        x, y = batch
                        # –ø–µ—Ä–µ–Ω–æ—Å–∏–º –Ω–∞ –≥–ø—É
                        x = x.to(device)
                        y = y.to(device)
                        # –ø—Ä–æ–≥–Ω–æ–∑–∏—Ä—É–µ–º
                        y_pred = net(x)
                        # loss = loss_fn(y_pred, y)
                        # –Ω–∞—Ö–æ–¥–∏–º accuracy –±–∞—Ç—á–∞ —Å —Ç–µ—Å—Ç–∞
                        accuracy.append(accuracy_score(y.detach().cpu().numpy(), np.argmax(y_pred.detach().cpu().numpy(), axis=1)))
                    # —É—Å—Ä–µ–¥–Ω—è–µ–º accuracy –≤—Å–µ—Ö –±–∞—Ç—á–µ–π –Ω–∞ —Ç–µ—Å—Ç–µ
                    accuracy = np.mean(np.array(accuracy))
                    # –µ—Å–ª–∏ —Å—Ç–∞–ª–æ –ª—É—á—à–µ - —Å–æ—Ö—Ä–∞–Ω—è–µ–º –Ω–∞ –¥–∏—Å–∫ –∏ –æ–±–Ω–æ–≤–ª—è–µ–º –ª—É—á—à—É—é –º–µ—Ç—Ä–∏–∫—É
                    if accuracy > best_accuracy:
                        print('New best model with test acc:', accuracy)
                        torch.save(net.state_dict(), './best_model.pt')
                        best_accuracy = accuracy

            # –≤—ã–≤–µ–¥–µ–º —Ç–µ–∫—É—â–∏–π loss
            running_loss += loss.item()
            # –≤—ã–≤–µ–¥–µ–º –∫–∞—á–µ—Å—Ç–≤–æ –∫–∞–∂–¥—ã–µ 500 –±–∞—Ç—á–µ–π
            if i % 500 == 499:
                print('[%d, %5d] loss: %.3f' %
                        (epoch + 1, i + 1, running_loss / 500))
                running_loss = 0.0

    print('–û–±—É—á–µ–Ω–∏–µ –∑–∞–∫–æ–Ω—á–µ–Ω–æ')
    return net
```
